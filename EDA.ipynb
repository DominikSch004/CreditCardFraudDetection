```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Data/creditcard.csv')
print(df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
df_missing_columns = (round(((df.isnull().sum()/len(df.index))*100),2).to_frame('null')).sort_values('null', ascending=False)
df_missing_columns
```

```python
classes = df["Class"].value_counts()
print(classes)
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
plt.title("Fraudulent vs. Non-fraudulent transactions")
sns.countplot(x='Class', data=df, ax=ax)
```

Distribution of classes over time

```python
fraud_df = df[df["Class"] == 1]
non_fraud_df = df[df["Class"] == 0]
```

```python
fig, ax = plt.subplots(figsize=(30, 15))
sns.distplot(fraud_df["Time"], hist=True, ax=ax, label="fraudulent", bins=10)
sns.distplot(non_fraud_df["Time"], hist=True, ax=ax, label="fraudulent", bins=10)
plt.title("transaction over time")
```

Distribution of classes over the amount of money

```python
fig, ax = plt.subplots(figsize=(30, 15))
sns.distplot(fraud_df["Amount"], hist=False, ax=ax, label="fraudulent")
sns.distplot(non_fraud_df["Amount"], hist=False, ax=ax, label="not fraudulent")
ax.set(xlabel='Transction Amount')
plt.legend()
plt.show()
```

```python
df.drop(columns=["Amount"], inplace=True)
```

Checking the Skewness

```python
columns = df.columns

k = 0
plt.figure(figsize=(30,28))
for col in columns:    
    k += 1
    plt.subplot(8, 5,k)    
    sns.distplot(df[col])
    plt.title(col+' '+str(df[col].skew()))
```
